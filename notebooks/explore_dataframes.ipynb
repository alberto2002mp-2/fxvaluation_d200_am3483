# DataFrame Exploration

This notebook defines helper functions to load the pre-built dataframes from
`src.data.load_excel_sheets` and then displays several of them to verify that
they look as expected.

```python
# Import Required Modules
import importlib
from typing import Dict
import sys
import pathlib

import pandas as pd

# ensure workspace root is on path so 'src' can be imported
root = pathlib.Path.cwd().parents[0]
if str(root) not in sys.path:
    sys.path.insert(0, str(root))

# we will import the loader module; reload later if needed
from src.data import load_excel_sheets

# ensure we have the latest version if the module is edited during the session
load_excel_sheets = importlib.reload(load_excel_sheets)

print("loader module imported:", load_excel_sheets)
```

```python
# Define Helper Functions to Load DataFrames

def get_sheet_dataframes(module) -> Dict[str, pd.DataFrame]:
    """Return a dictionary of attribute name to DataFrame found in `module`.

    The loader module exposes variables like ``fx_df`` etc.  We also support
    accidentally exposing other DataFrame objects; this helper filters by type.
    """
    result: Dict[str, pd.DataFrame] = {}
    for name in dir(module):
        if name.endswith("_df"):
            attr = getattr(module, name)
            if isinstance(attr, pd.DataFrame):
                result[name] = attr
    return result


def load_all(excel_path: str | pathlib.Path = None) -> Dict[str, pd.DataFrame]:
    """Reload the module and return all dataframes afresh.
    
    Parameters
    ----------
    excel_path : str or pathlib.Path, optional
        Path to the Excel file. If not provided, uses data/rawdata.xlsx relative
        to the workspace root.
    """
    if excel_path is None:
        # compute path relative to the workspace root
        excel_path = root / "data" / "rawdata.xlsx"
    
    # Call the loader's main function with the correct path
    from src.data.load_excel_sheets import load_all_sheets
    load_all_sheets(excel_path)
    
    # Now reload and extract the module-level dataframes
    mod = importlib.reload(load_excel_sheets)
    return get_sheet_dataframes(mod)
```

```python
# Load DataFrames from Python File

frames = load_all()
frames
```

```python
# Verify Loaded DataFrames

for name, df in frames.items():
    print(f"{name}: {df.shape}")
    display(df.head())
```
