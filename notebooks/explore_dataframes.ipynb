```python
# Define Helper Functions to Load DataFrames

def get_sheet_dataframes(module) -> Dict[str, pd.DataFrame]:
    """Return a dictionary of attribute name to DataFrame found in `module`.

    The loader module exposes variables like ``fx_df`` etc.  We extract those
    by filtering for names that end with "_df" and are DataFrame objects.
    """
    result: Dict[str, pd.DataFrame] = {}
    for name in dir(module):
        if name.endswith("_df"):
            attr = getattr(module, name)
            if isinstance(attr, pd.DataFrame):
                result[name] = attr
    return result


def load_all() -> Dict[str, pd.DataFrame]:
    """Reload the loader module and return all dataframes.
    
    The loader automatically resolves paths from config.py, so no path argument
    is needed here. This function simply triggers a reload and extracts the
    module-level dataframe variables.
    """
    mod = importlib.reload(load_excel_sheets)
    return get_sheet_dataframes(mod)
```

```python
import numpy as np

# --- 1. Setup: constituent columns and weights ---
cols = [
    "EURUSD Curncy", "USDJPY Curncy", "USDCAD Curncy", "GBPUSD Curncy", "USDMXN Curncy",
    "USDCNH Curncy", "USDCHF Curncy", "AUDUSD Curncy", "USDKRW Curncy", "USDINR Curncy",
    "USDSGD Curncy", "USDTWD Curncy"
]
base_weights = np.array([0.2947, 0.1238, 0.1165, 0.1027, 0.0962, 0.0700, 0.0447, 0.0439, 0.0316, 0.0283, 0.0261, 0.0215])
w = pd.Series(base_weights, index=cols)

# --- 2. Reindex all constituent prices to BBDXY's valid index ---
bbdxy_index = fx_df["BBDXY Curncy"].dropna().index
px = fx_df[cols].reindex(bbdxy_index)

# --- 3. Oriented log returns matrix ---
signs = np.where([c.startswith("USD") for c in cols], 1, -1)
log_px = np.log(px)
R_df = (signs * log_px.diff())
R_df = R_df.fillna(0.0)  # fill missing returns with 0.0

# --- 4. Exclusion weight matrix ---
exclusions = ["EURUSD Curncy", "USDJPY Curncy", "USDCHF Curncy", "GBPUSD Curncy", "AUDUSD Curncy", "USDCAD Curncy"]
K = len(exclusions)
W_ex = np.tile(w.values, (K,1)).T  # shape (12,K)
for k, excl in enumerate(exclusions):
    W_ex[cols.index(excl), k] = 0  # set excluded weight to 0
W_ex = pd.DataFrame(W_ex, index=cols, columns=exclusions)

# --- 5. CNH regime split ---
CNH_START = pd.Timestamp("2010-08-23")
pre_mask = R_df.index < CNH_START
post_mask = R_df.index >= CNH_START

def normalize_weights(W, cnh_active):
    Wn = W.copy()
    if not cnh_active:
        Wn.loc["USDCNH Curncy", :] = 0  # remove CNH
    Wn = Wn.div(Wn.sum(axis=0), axis=1)
    return Wn

W_pre = normalize_weights(W_ex, cnh_active=False)
W_post = normalize_weights(W_ex, cnh_active=True)

# --- 6. Compute weighted returns (matrix multiplication) ---
R_pre = R_df.loc[pre_mask]
R_post = R_df.loc[post_mask]

r_pre_df = R_pre.dot(W_pre)
r_post_df = R_post.dot(W_post)

# --- 7. Concatenate, compute index levels ---
r_all = pd.concat([r_pre_df, r_post_df]).sort_index()
base_level = fx_df["BBDXY Curncy"].dropna().iloc[0]
idx_all = base_level * np.exp(r_all.cumsum())

# --- 8. Name columns as required ---
colnames = [
    "BBDXY_ex_EUR", "BBDXY_ex_JPY", "BBDXY_ex_CHF",
    "BBDXY_ex_GBP", "BBDXY_ex_AUD", "BBDXY_ex_CAD"
]
idx_all.columns = colnames

# --- 9. # Concatenate the new synthetic indices with the original fx_df
fx_newdf = pd.concat([fx_df, idx_all], axis=1)

# Optionally: correlation with official BBDXY
print("\nCorrelation with BBDXY Curncy:")
print(fx_newdf.corrwith(fx_df["BBDXY Curncy"]))
```

```python
# --- 1. Helper functions ---

def log_return(series):
    """Compute log returns, preserving index."""
    return np.log(series / series.shift(1))

def daily_change(series):
    """Compute daily value change, preserving index."""
    return series.diff()

def transform_series(series, col_name):
    """
    Apply the correct transformation to a series based on its name.
    - daily_change if name endswith BLC Curncy, BLC2 Curncy, YR Index, YR Corp, or startswith RR2YH
    - log_return otherwise
    """
    if (
        col_name.endswith("BLC Curncy")
        or col_name.endswith("BLC2 Curncy")
        or col_name.endswith("YR Index")
        or col_name.endswith("YR Corp")
        or col_name.startswith("RR2YH")
    ):
        return daily_change(series)
    else:
        return log_return(series)

def build_driver(df, spec):
    """
    Build a driver series from a DataFrame and a driver spec.
    - If type == 'single', transform the specified column.
    - If type == 'spread', transform left and right, then subtract.
    Returns None if required columns are missing.
    """
    if spec["type"] == "single":
        col = spec["col"]
        if col in df.columns:
            return transform_series(df[col], col)
        else:
            return None
    elif spec["type"] == "spread":
        left, right = spec["left"], spec["right"]
        if left in df.columns and right in df.columns:
            return transform_series(df[left], left) - transform_series(df[right], right)
        else:
            return None
    else:
        raise ValueError(f"Unknown driver spec type: {spec['type']}")

def process_dataframe(df, driver_specs):
    """
    Process a DataFrame according to driver_specs.
    - Forward-fill missing values.
    - For each driver, build the series if possible.
    - Rename commodity columns if commodity_rename is provided.
    - Drop the first row (NaN from diff/log).
    - Return processed DataFrame with only driver columns.
    """
    df_ffill = df.ffill()
    out = {}

    # Build drivers from spec
    for driver, spec in driver_specs.items():
        series = build_driver(df_ffill, spec)
        if series is not None:
            out[driver] = series
    if not out:
        return pd.DataFrame(index=df_ffill.index)  # empty, but preserves index
    result = pd.DataFrame(out)
    # Drop the first row (NaN from diff/log)
    result = result.iloc[1:]
    return result


```

```python
# --- Rolling 1-year z-score transformation for processed dataframes ---

window = 22
df_1yz_map = {}

for key, df in processed_df_map.items():
    rolling_mean = df.rolling(window=window, min_periods=window).mean()
    rolling_std = df.rolling(window=window, min_periods=window).std()

    z = (df - rolling_mean) / rolling_std
    z = z.replace([np.inf, -np.inf], np.nan).dropna(how="all")

    new_key = key.replace("_df_processed", "_df_1yz")
    df_1yz_map[new_key] = z

    globals()[new_key] = z  # optional, only if you want standalone dataframe names

first_key = next(iter(df_1yz_map))
print(f"{first_key} shape after rolling z-score transformation: {df_1yz_map[first_key].shape}")
for name, df in df_1yz_map.items():
    print(f"{name}: {df.shape}, {df.index.min().date()} → {df.index.max().date()}")
```

```python
import matplotlib.pyplot as plt

def plot_driver(df_name, column_name):
    """
    Plot a time series from a specified dataframe and column in df_1yz_map.
    Args:
        df_name (str): Name of the dataframe in df_1yz_map (e.g., 'aud_df_1yz').
        column_name (str): Name of the column (driver) to plot (e.g., '2y yield').
    Returns:
        matplotlib.figure.Figure: The matplotlib figure object.
    Raises:
        KeyError: If the dataframe or column does not exist.
    """
    # Validate dataframe existence
    if df_name not in df_1yz_map:
        raise KeyError(f"DataFrame '{df_name}' not found in df_1yz_map.")
    df = df_1yz_map[df_name]
    # Validate column existence
    if column_name not in df.columns:
        raise KeyError(f"Column '{column_name}' not found in DataFrame '{df_name}'.")
    series = df[column_name]
    # Plotting
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(series.index, series.values, label=column_name)
    ax.set_title(f"{column_name} — {df_name}", fontsize=16)
    ax.set_xlabel("Date", fontsize=12)
    ax.set_ylabel("Z-Score", fontsize=12)
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)
    ax.legend()
    plt.tight_layout()
    return fig

# Example usage:
plot_driver("eur_df_1yz", "EURUSD")
```
