# DataFrame Exploration

This notebook defines helper functions to load the pre-built dataframes from
`src.data.load_excel_sheets` and then displays several of them to verify that
they look as expected.

```python
# Import Required Modules
import importlib
from typing import Dict
import sys
import pathlib

import pandas as pd

# Ensure workspace root is on path so 'src' and 'config' can be imported
root = pathlib.Path.cwd().parents[0]
if str(root) not in sys.path:
    sys.path.insert(0, str(root))

# Import the loader module
from src.data import load_excel_sheets

# Reload to get the latest version if the module was edited during the session
load_excel_sheets = importlib.reload(load_excel_sheets)

print("✓ Loader module imported successfully from", load_excel_sheets.__file__)
```

```python
# Define Helper Functions to Load DataFrames

def get_sheet_dataframes(module) -> Dict[str, pd.DataFrame]:
    """Return a dictionary of attribute name to DataFrame found in `module`.

    The loader module exposes variables like ``fx_df`` etc.  We extract those
    by filtering for names that end with "_df" and are DataFrame objects.
    """
    result: Dict[str, pd.DataFrame] = {}
    for name in dir(module):
        if name.endswith("_df"):
            attr = getattr(module, name)
            if isinstance(attr, pd.DataFrame):
                result[name] = attr
    return result


def load_all() -> Dict[str, pd.DataFrame]:
    """Reload the loader module and return all dataframes.
    
    The loader automatically resolves paths from config.py, so no path argument
    is needed here. This function simply triggers a reload and extracts the
    module-level dataframe variables.
    """
    mod = importlib.reload(load_excel_sheets)
    return get_sheet_dataframes(mod)
```

```python
# Load DataFrames from Python File

frames = load_all()
frames
```

```python
# Verify Loaded DataFrames

for name, df in frames.items():
    print(f"{name}: {df.shape}")
    display(df.head())
```

```python
  # Template: Drop Columns from DataFrames

def drop_columns(df: pd.DataFrame, columns_to_drop: list[str]) -> pd.DataFrame:
    """Drop specified columns from a dataframe.
    
    Parameters
    ----------
    df : pd.DataFrame
        The dataframe to filter.
    columns_to_drop : list[str]
        List of column names to remove.
    
    Returns
    -------
    pd.DataFrame
         dataframe with specified columns removed.
        
    Raises
    ------
    ValueError
        If any specified column doesn't exist in the dataframe.
    """
    missing = [col for col in columns_to_drop if col not in df.columns]
    if missing:
        raise ValueError(f"Columns not found in dataframe: {missing}")
    
    return df.drop(columns=columns_to_drop)


# Example: Drop columns from sek_df
sek_df = drop_columns(
    frames["sek_df"],
    columns_to_drop=[
        "G0021 3M1M BLC2 Curncy",
        "G0021 6M1M BLC2 Curncy",
        "G0021 1Y1M BLC2 Curncy"
    ]
)

print(f"\nRemaining columns in sek_df:\n{list(sek_df.columns)}")
print(f"Original sek_df shape: {frames['sek_df'].shape}")
print(f" sek_df shape: {sek_df.shape}")



# Example: Drop columns from nok_df
nok_df = drop_columns(
    frames["nok_df"],
    columns_to_drop=[
        "G0078 3M1M BLC2 Curncy",
        "G0078 6M1M BLC2 Curncy",
        "G0078 1Y1M BLC2 Curncy"
    ]
)

print(f"\nRemaining columns in nok_df:\n{list(nok_df.columns)}")
print(f"Original nok_df shape: {frames['nok_df'].shape}")
print(f" nok_df shape: {nok_df.shape}")

# Example: Drop columns from aud_df
aud_df = drop_columns(
    frames["aud_df"],
    columns_to_drop=[
        "G0001 3M1M BLC2 Curncy",
        "G0001 6M1M BLC2 Curncy",
        "G0001 1Y1M BLC2 Curncy"
    ]
)

print(f"\nRemaining columns in aud_df:\n{list(aud_df.columns)}")
print(f"Original aud_df shape: {frames['aud_df'].shape}")
print(f" aud_df shape: {aud_df.shape}")


# Example: Drop columns from nzd_df
nzd_df = drop_columns(
    frames["nzd_df"],
    columns_to_drop=[
        "G0049 3M1M BLC2 Curncy",
        "G0049 6M1M BLC2 Curncy",
        "G0049 1Y1M BLC2 Curncy"
    ]
)

print(f"\nRemaining columns in nzd_df:\n{list(nzd_df.columns)}")
print(f"Original nzd_df shape: {frames['nzd_df'].shape}")
print(f" nzd_df shape: {nzd_df.shape}")


# Example: Drop columns from chf_df
chf_df = drop_columns(
    frames["chf_df"],
    columns_to_drop=[
        "G0082 3M1M BLC2 Curncy",
        "G0082 6M1M BLC2 Curncy",
        "G0082 1Y1M BLC2 Curncy"
    ]
)

print(f"\nRemaining columns in chf_df:\n{list(chf_df.columns)}")
print(f"Original chf_df shape: {frames['chf_df'].shape}")
print(f" chf_df shape: {chf_df.shape}")


# Example: Drop columns from jpy_df
jpy_df = drop_columns(
    frames["jpy_df"],
    columns_to_drop=[
        "G0018 3M1M BLC2 Curncy",
        "G0018 6M1M BLC2 Curncy",
        "G0018 1Y1M BLC2 Curncy"
    ]
)

print(f"\nRemaining columns in jpy_df:\n{list(jpy_df.columns)}")
print(f"Original jpy_df shape: {frames['jpy_df'].shape}")
print(f" jpy_df shape: {jpy_df.shape}")


fx_df = frames["fx_df"]
print(f"\nColumns in fx_df:\n{list(fx_df.columns)}")
print(f" jpy_df shape: {fx_df.shape}")

usd_df = frames["usd_df"]
print(f"\nColumns in usd_df:\n{list(usd_df.columns)}")
print(f" usd_df shape: {usd_df.shape}")

eur_df = frames["eur_df"]
print(f"\nColumns in eur_df:\n{list(eur_df.columns)}")
print(f" eur_df shape: {eur_df.shape}")
gbp_df = frames["gbp_df"]
print(f"\nColumns in gbp_df:\n{list(gbp_df.columns)}")
print(f" gbp_df shape: {gbp_df.shape}")

cad_df = frames["cad_df"]
print(f"\nColumns in cad_df:\n{list(cad_df.columns)}")
print(f" cad_df shape: {cad_df.shape}")
```
